```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np 


pixels = []
with open("LOW_res.txt", "r", encoding="utf-8") as f:
    for n in f:
        n.strip('\n')
        i = int(n)
        pixels.append(i)
        
a = np.array(pixels)
a.shape = (10, 10)
plt.imshow(a, norm=LogNorm())
plt.colorbar()
plt.savefig('SUBMIT_Low.png', dpi=900)
plt.show()
```
