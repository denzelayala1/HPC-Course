```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np # We are going use numpy later on


pixels = []
with open("low_res_mesh.txt", "r", encoding="utf-8") as f:
    for n in f:
        n.strip('\n')
        i = int(n)
        pixels.append(i)
        
a = np.array(pixels)
a.shape = (10, 10)
plt.imshow(a, norm=LogNorm())
plt.colorbar()
plt.show()
```
