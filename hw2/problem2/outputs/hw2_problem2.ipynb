```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np 
```

```python
xvals = np.linspace(-2,2,8)
yvals = np.linspace(-1,1,4)
```

```python
pixels = []
with open("Cores-2__res-1000000__Rank-0__Mandelbrot_set.txt", "r", encoding="utf-8") as f:
    for n in f:
        n.strip('\n')
        i = int(n)
        pixels.append(i)
        
a = np.array(pixels)
a.shape = (1000, 1000)
plt.imshow(a, norm=LogNorm())
plt.colorbar()
plt.savefig('partA_Full.png', dpi=2400)
plt.show()
```
