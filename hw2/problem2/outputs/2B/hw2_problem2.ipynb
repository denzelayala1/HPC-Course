```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np 

pixels = []
with open("res-1000000__Cores-2__Rank-0__Mandelbrot_set.txt", "r", encoding="utf-8") as f:
    for n in f:
        n.strip('\n')
        i = int(n)
        pixels.append(i)
        
a = np.array(pixels)
a.shape = (1000, 1000)
plt.imshow(a, norm=LogNorm())
plt.colorbar()
plt.savefig('res-1000000__Cores-2__Rank-0__Mandelbrot_set.png', dpi=2400)
plt.show()
```
